```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis,y_axis)
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston', color ="blue",linewidth=2,marker="*")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
#add a grid
plt.grid()
# Add the legend.
plt.legend()
```

```python
fig, ax = plt.subplots(linewidth=2)

# Create a title.
ax.set_title("PyBer Fare by Month")
#create labels
ax.set_xlabel("Months")
ax.set_ylabel("Fares ($)")

#add grid
ax.grid()
#set y limits
ax.set_ylim(0, 45)

ax.plot(x_axis, y_axis,label="Boston",color = "green", marker="^")
#add legend
ax.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(y_axis,x_axis, "o", color = "red", label = "Chicago")

plt.xlabel("fare ($)")
plt.ylabel("Date")

#Create Title
plt.title("Pyber Fare by Month")

#add legend
plt.legend()

#add x limit
plt.xlim(0,50)

#inverse y axix
plt.gca().invert_yaxis()
```
